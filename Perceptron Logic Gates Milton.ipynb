```python
from sklearn.linear_model import Perceptron
import matplotlib.pyplot as plt
import numpy as np
from itertools import product
```

## Representing an AND gate

```python
data = [[0,0],[0,1],[1,0],[1,1]]
labels = [0,0,0,1]
```

```python
x = [point[0] for point in data]
y = [point[1] for point in data]
plt.scatter(x, y, c=labels)

plt.show()
```

```python
#max_iter sets the number of times the perceptron loops through the training data. 
#(The default is 1000, so we're cutting the training pretty short) 
classifier = Perceptron(max_iter=40,random_state=22)
classifier.fit(data,labels)

#Note that it is pretty unusual to train and test on the same dataset. 
#Since there are only 4 possible inputs to AND, we train on every possible input and test on those same points.
print(classifier.score(data,labels))
```

## Representing an XOR gate

```python
datax = [[0,0],[0,1],[1,0],[1,1]]
labelsx = [0,1,1,0]

x1 = [point[0] for point in datax]
y1 = [point[1] for point in datax]
plt.scatter(x1, y1, c=labelsx)

plt.show()

#max_iter sets the number of times the perceptron loops through the training data. 
#(The default is 1000, so we're cutting the training pretty short) 
classifier1 = Perceptron(max_iter=40,random_state=22)
classifier1.fit(datax,labelsx)

#Note that it is pretty unusual to train and test on the same dataset. 
#Since there are only 4 possible inputs to AND, we train on every possible input and test on those same points.
print(classifier1.score(datax,labelsx))
```

## Representing an OR gate

```python
datao = [[0,0],[0,1],[1,0],[1,1]]
labelso = [0,1,1,1]

x2 = [point[0] for point in datao]
y2 = [point[1] for point in datao]
plt.scatter(x2, y2, c=labelso)

plt.show()

#max_iter sets the number of times the perceptron loops through the training data. 
#(The default is 1000, so we're cutting the training pretty short) 
classifier2 = Perceptron(max_iter=40,random_state=22)
classifier2.fit(datao,labelso)

#Note that it is pretty unusual to train and test on the same dataset. 
#Since there are only 4 possible inputs to AND, we train on every possible input and test on those same points.
print(classifier2.score(datao,labelso))
```

## Visualizing the Perceptron

```python
print(classifier.decision_function([[0, 0], [1, 1], [0.5, 0.5]]))
```

```python
x_values = np.linspace(0, 1, 100)
y_values = np.linspace(0, 1, 100)

point_grid = list(product(x_values,y_values))

distances = classifier2.decision_function(point_grid)
abs_distances = [abs(i) for i in distances]
distances_matrix = np.reshape(abs_distances, (100,100))

heatmap = plt.pcolormesh(x_values,y_values,distances_matrix)
plt.colorbar(heatmap)
plt.scatter(x2, y2, c=labelso)
plt.show()
```
